```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from scipy.constants import constants
matplotlib.rcParams['figure.dpi'] = 500
matplotlib.rcParams['xtick.direction'] = 'in' 
```

```python
def file(filename):
    F = open(filename)
    F = F.readlines()
    result = []
    for line in F:
        if '0' <= line[0] <='9':
            result.append([float(line[0:6]), float(line[-7:-1])])
    return np.array(result)

def file2(filename):
    F = open(filename)
    F = F.readlines()
    result = []
    for line in F:
        if len(line)>1 and '0' <= line[-2] <='9':
            result.append(float(line[-6:-1]))
    return np.array(result)


chinese = matplotlib.font_manager.FontProperties(fname='C:\Windows\Fonts\simkai.ttf',size=15)
res1 = file('C:/Users/ASUS/OneDrive/桌面/workingfile/氢光谱/βγδ(1).gd6')
res2 = file('C:/Users/ASUS/OneDrive/桌面/workingfile/氢光谱/第一次定标.gd6')
res3 = file('C:/Users/ASUS/OneDrive/桌面/workingfile/氢光谱/第二次定标.gd6')
res4 = file('C:/Users/ASUS/OneDrive/桌面/workingfile/氢光谱/alpha.gd6')


length = np.append(res1[:,0],res4[133:,0])
res = np.append(res1[:,1]/2,res4[133:,1])
peaks = []
for idx, l in enumerate(res):
   if (20 < idx < 3900) and l>140 and \
   l>=max(res[idx-5: idx+1]) and l<=np.max(res[idx-5:idx+1])+200 \
       and l>=max(res[idx: idx+5]) and l<=np.max(res[idx+1:idx+5])+200:
       peaks.append(idx)
fig, (ax1,ax2,ax3)= plt.subplots(3, 1, figsize=(8, 5))
ax1.plot(res1[:,0], res1[:,1]/2, lw=0.8,color='green', label=r"第一次测量")
ax1.plot(res2[:,0], res2[:,1]/2, lw=0.8,color='blue',label=r"第一次定标")
ax1.grid()

ax1.legend(prop=chinese)
ax1.set_ylim([0, 2000])
ax2.plot(res3[:,0], res3[:,1], lw=0.8,color='blue',label=r"第二次定标")
ax2.plot(res4[:,0], res4[:,1], lw=0.8,color='red',label=r"第二次测量")
ax2.set_ylim([0, max(max(res3[:,1]), max(res4[:, 1]))+3])
ax2.grid()

ax2.legend(prop=chinese)
ax2.set_ylim([0, 2000])

ax3.plot(np.append(res2[:,0],res3[133:,0]), np.append(res2[:,1]/2,res3[133:,1]),
         lw=0.8,color='blue',label=r"定标")
ax3.plot(length, res,
         lw=0.8,color='red',label=r"氢光谱",zorder= 5)
ax3.scatter(length[peaks], res[peaks],marker='x', zorder=40,s=20,linewidth=1,c='black')

ax3.annotate(r"$H_\epsilon$({:.2f}nm)".format(length[peaks[0]]), fontsize=10,
             family="serif", xy= (length[peaks[0]], res[peaks[0]]), xycoords="data",
             xytext=(-8,52.6), textcoords="offset points",
           arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.3"))
ax3.annotate(r"$H_\delta$({:.2f}nm)".format(length[peaks[1]]), fontsize=10,
             family="serif", xy= (length[peaks[1]], res[peaks[1]]), xycoords="data",
             xytext=(-13,19), textcoords="offset points",
           arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.5"))
ax3.annotate(r"$H_\gamma$({:.2f}nm)".format(length[peaks[2]]), fontsize=10,
             family="serif", xy= (length[peaks[2]], res[peaks[2]]), xycoords="data",
             xytext=(19,11.5), textcoords="offset points",
           arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=-.3"))

ax3.annotate(r"$H_\beta$({:.2f}nm)".format(length[peaks[3]]), fontsize=10,
             family="serif", xy= (length[peaks[3]], res[peaks[3]]), xycoords="data",
             xytext=(20,11.7), textcoords="offset points",
           arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=-.3"))
ax3.annotate(r"$H_\alpha$({:.2f}nm)".format(length[peaks[4]]), fontsize=10,
             family="serif", xy= (length[peaks[4]], res[peaks[4]]), xycoords="data",
             xytext=(-80,14), textcoords="offset points",
           arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.1"))

ax3.set_xlim([390, 750])
ax3.set_ylim([0, 3000])

ax3.legend(prop=chinese)
plt.tight_layout()
fig.savefig(r'C:/Users/ASUS/OneDrive/桌面/氢光谱1.jpeg', dpi=400)
```

```python
R_H = constants.Rydberg*(constants.proton_mass/(constants.electron_mass+constants.proton_mass))
E_1 =  -constants.h*constants.c*R_H/(constants.e)
e = []
for i in [1,2,3]:
    k = []
    for j in range(i+1,i+7):
        k.append([E_1/i**2, E_1/j**2, 10**9*constants.h*constants.c/(constants.e*(E_1/j**2-E_1/i**2))])
    e.append(k)
e = np.array(e)
lab = ['α','β','γ','δ','ε' ]
fig, ax = plt.subplots(1,1, figsize=(6,5))
ax.set_xticks([])
ax.set_xticklabels([])
ax.set_xlim([80, 2200])
ax.hlines(0,80, 2200,linewidth=.7,color='black')
ax.annotate("$\infty$", xy=(500, E_1/i**2), xytext=(69.5,-0.2))
ax.annotate("$n$", xy=(500, E_1/i**2), xytext=(71,0.6))
ax.annotate("0", xy=(500, E_1/i**2), xytext=(2380,-0.2))
ax.annotate("$E/\mathrm{eV}$", xy=(500, E_1/i**2), xytext=(2300,+0.6))
for i in [1,2,3,4,5,6,7,8,9]:
    ax.hlines(E_1/i**2,80,2200,linewidth=.7,color='black')
    if i <=4:
        ax.annotate("{}".format(i), xy=(500, E_1/i**2), xytext=(70,E_1/i**2-0.2))
        ax.annotate("{:.2f}".format(E_1/i**2), xy=(500, E_1/i**2), xytext=(2300,E_1/i**2-0.2))
for idx1, i in enumerate(e):
    for idx2, j in enumerate(i):
        ax.annotate("", xy=(j[2], j[0]-0.1), xytext=(j[2], j[1]+0.1),arrowprops=dict(arrowstyle="->"))
        if idx1 == 1 and (0<=idx2<=4):
            ax.annotate("$H_{}$".format(lab[idx2]), xy=(j[2]-12,  (j[0]+j[1])/2-0.5+0.1*idx2), xytext=(j[2]+520+140*(4-idx2), (j[0]+j[1])/2-0.56+.03*idx2**2),arrowprops=dict(arrowstyle="-|>"))
def forward(arr):
    result = np.zeros_like(arr)
    for idx, ele in enumerate(arr):
        if ele <300:
            result[idx] = 5*ele
        else:
            result[idx] =  1000 +1000*np.log10(ele/300)
    return result
def rever(arr):
    result = np.zeros_like(arr)
    for idx, ele in enumerate(arr):
        if ele < 300:
            result[idx] = ele/5
        else:
            result[idx] =  300*10**((ele-1000)/1000)
    return result
axi = matplotlib.scale.FuncScale(ax,(forward, rever))
ax.set_xscale(axi)
ax.set_xticks([100, 150, 200, 500, 1000, 2000]) 
ax.set_xticklabels([100, 150, 200, 500, 1000, 2000], fontsize=10)
ax.set_yticks([]) 
ax.tick_params(axis=u'y', which=u'both',length=0)
ax.set_title("$\lambda/\mathrm{nm}$",fontsize=10, y=-0.13)

ax4 =  fig.add_axes([0.496, 0.484, 0.147, 0.19], facecolor="#f5f5f5")
ax4.set_xscale(axi)
plt.xticks([])
ax4.set_ylim([500,3000])
ax4.set_yticks([]) 
ax4.plot(length, 3000-res,
         lw=0.8,color='red',label=r"氢光谱",zorder= 5)
ax4.scatter(length[peaks], 3000-res[peaks],marker='x', zorder=40,s=20,linewidth=1,c='black')
ax4.set_title("实验结果",fontproperties=chinese,y=-0.05,fontsize=10)
fig.savefig(r'C:/Users/ASUS/OneDrive/桌面/能级图.jpg',dpi=400)
```
