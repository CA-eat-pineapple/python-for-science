```python
colors=['red', 'blue','green', 'orange']
labels = [r'幅度15，无软管，谐振频率：',r'幅度25，无软管，谐振频率：',r'幅度15,1个软管，谐振频率：',r'幅度15,2个软管，谐振频率：']
```

```python
fig,ax=plt.subplots(1,1,figsize=(7,5))
ax.grid(zorder=-100)
for i in range(4):
    x = data[2*i]
    peak_x = x[3]
    y = data[2*i+1]
    peak_y = y[3]
    test_x = np.concatenate((np.linspace(x[0],x[1],10), \
                             np.linspace(x[1],x[2],10), np.arange(x[2], x[-4],0.01),np.linspace(x[-4],x[-3],4), np.linspace(x[-2], x[-1],10)))
    ipo2=spi.splrep(x,y,k=1) #样本点导入，生成参数
    iy2=spi.splev(test_x,ipo2)
    ax.scatter(x, y,color=colors[i])
    ax.plot(test_x,iy2,label=labels[i]+str(peak_x)+'$\;\mathrm{Hz}$',zorder=1,color=colors[i])
    ax.set_ylim([0,0.15])
    ax.set_ylabel('音叉振动振幅$A/\mathrm{mm}$',fontsize=14)
    ax.set_xlabel('频率/$ \mathrm{Hz}$',fontsize=14)
    ax.legend(prop=chinese)
```
